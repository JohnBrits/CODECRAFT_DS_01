```python
import pandas as pd 
import numpy as np
df = pd.read_csv('C:\\Users\\britt\\Downloads\\Salary_Data.csv')
 
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming the dataset contains 'Age' and 'Gender' columns
# If the columns have different names, adjust accordingly

# Display age distribution using histogram
plt.figure(figsize=(12, 5))

# Age histogram
plt.subplot(1, 2, 1)
plt.hist(df['Age'], bins=10, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Gender bar chart
plt.subplot(1, 2, 2)
gender_counts = df['Gender'].value_counts()
plt.bar(gender_counts.index, gender_counts.values)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')

plt.tight_layout()
plt.show()

# Alternative with seaborn for more aesthetic plots
plt.figure(figsize=(12, 5))

# Age histogram with seaborn
plt.subplot(1, 2, 1)
sns.histplot(data=df, x='Age', kde=True)
plt.title('Age Distribution')

# Gender count plot
plt.subplot(1, 2, 2)
sns.countplot(data=df, x='Gender')
plt.title('Gender Distribution')

plt.tight_layout()
plt.show()

# If you want to see age distribution by gender
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', hue='Gender', multiple='stack')
plt.title('Distribution of Age in Gender Classification ')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()


```
